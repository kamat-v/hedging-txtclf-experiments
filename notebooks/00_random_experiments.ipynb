```python
# load parquet
df = pd.read_parquet("data/processed/train.parquet")   # change path if needed

labels = df["label"]

# count 0/1
counts = labels.value_counts().sort_index()

plt.figure(figsize=(6,4))
plt.bar(counts.index, counts.values)
plt.xticks([0,1])
plt.yticks([])
#plt.title("Training Data Imbalance")
plt.show()
```
